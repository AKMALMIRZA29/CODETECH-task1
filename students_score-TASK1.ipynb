drop unused column

```python
df= df.drop('Unnamed: 0', axis = 1)
```

```python
df.head()
```

gender distribution

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df, x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("gender distribution")
plt.show()
```

from above chart we have analysed that : the number of females in the data is more that the number of males

students scores by parents education 

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title('impact on students score by their parents education ')
plt.show()
```

from above chart we conclude that the education of parents have a good impact on their scores

students scores by parents marital status

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title('impact on students score by their parents martial status ')

plt.show()
```

from above chart we conclude that the martial status of parents have a neglegible impact on students scores

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l=["groupA", "groupB", "groupC", "groupD", "groupE"]
mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

print(mylist)
plt.title("Distribution of ethnic groups")
plt.pie(mylist,labels = l, autopct = "%1.2f%%")
plt.show()


```
